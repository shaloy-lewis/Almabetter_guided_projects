```python
def conv_str_to_list(text):
  return eval(text)
```

```python
imdb_df['cast'] = imdb_df['cast'].apply(conv_str_to_list)
```

```python
imdb_df['crew'] = imdb_df['crew'].apply(conv_str_to_list)
imdb_df['genres'] = imdb_df['genres'].apply(conv_str_to_list)
```

1. Which movie made the highest profit? Who were its producer and director? Identify the actors in that film.

```python
imdb_df.describe()
```

```python
imdb_df.loc[imdb_df['budget']<1000, 'budget'] = imdb_df.loc[imdb_df['budget']>=1000, 'budget'].median()

imdb_df.loc[imdb_df['revenue']<1000, 'revenue'] = imdb_df.loc[imdb_df['revenue']>=1000, 'revenue'].median()
```

```python
imdb_df['profit'] = imdb_df['revenue']-imdb_df['budget']
```

```python
imdb_df.head()
```

```python
max_profit = imdb_df['profit'].max()
max_profit
```

```python
max_profit_movie_df = imdb_df[imdb_df['profit']==max_profit]
max_profit_movie_df['title']
```

```python
crew_list = max_profit_movie_df.iloc[0]['crew']
crew_list
```

```python
producer_list = []
director_list = []

for elem in crew_list:
  if elem['job'] == 'Producer':
    producer_list.append(elem['name'])
  if elem['job'] == 'Director':
    director_list.append(elem['name'])
```

```python
producer_list, director_list
```

```python
cast_list = max_profit_movie_df.iloc[0]['cast']
cast_list
```

```python
actor_list=[]

for elem in cast_list:
  actor_list.append(elem['name'])

actor_list
```

2. This data has information about movies made in different languages. Which language has the highest average ROI (return on investment)?

```python
imdb_df['roi'] = 100*imdb_df['profit']/imdb_df['budget']
imdb_df.head()
```

```python
roi_lang_df = imdb_df.groupby('original_language')['roi'].mean().reset_index()
roi_lang_df
```

```python
max_roi_lan = roi_lang_df[roi_lang_df['roi']==roi_lang_df['roi'].max()]['original_language']
max_roi_lan
```

3. Find out the unique genres of movies in this dataset.

```python
genres_list = []

for index,row in imdb_df.iterrows():
  genre_list = row['genres']
  for elem in genre_list:
    genres_list.append(elem['name'])

```

```python
unique_genres = set(genres_list)
unique_genres
```

4. Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
def get_producers_and_directors(crew_list):
  try:
    producer_list = []
    director_list = []
    for elem in crew_list:
      if elem['job'] == 'Producer':
        producer_list.append(elem['name'])
      if elem['job'] == 'Director':
        director_list.append(elem['name'])
      return producer_list, director_list
  except:
    return [], []
```

```python
imdb_df[['producers','directors']] = imdb_df.apply(lambda x: get_producers_and_directors(x['crew']),axis=1,result_type='expand')
```
